```python
#11 December 2025

fname = input("Enter file name: ")
fh = open(fname)


total = 0
count = 0


for line in fh:
    if not line.startswith("X-DSPAM-Confidence:"):
        continue
    colon = line.find(':')
    total += float(line[(colon + 1): len(line)].rstrip())
    count += 1
   
print('Average spam confidence: ' + str(total/count))
```

```python
#16 December 2025

fname = input("Enter file name: ")
fh = open(fname)
count = 0
lst = list()

for line in fh:
    if line.startswith('From '):
        words = line.split()
        print(words[1])
        count+=1

for line in fh:
    pieces = line.split() 
    for piece in pieces:
        if piece not in lst:
            lst.append(piece)
            
lst.sort()                

print("There were", count, "lines in the file with From as the first word")
print(lst)
```

```python
#18 December 2025

fname = input("Enter file:")
fh = open(fname)

counts = dict()

for line in fh:
    if line.startswith('From '):
        from_line = line.split()
        name = from_line[1]
        counts[name] = counts.get(name, 0) + 1

max_num = None 
max_word = None 

for word,num in counts.items():
    if max_num is None or num > max_num:
        max_num = num 
        max_word = word 
        
print(max_word,max_num)
```
